```python
%matplotlib inline
import pylab as plt
import numpy as np

g_one = 100 + 15 * np.random.randn(10000)
g_two = 85 + 5 * np.random.randn(10000)
```

```python
# please show a histogram with g_one and g_two with labels and title
```

```python
fig, ax = plt.subplots()
plt.hist(g_one, label = 'g_one')
plt.hist(g_two, label = 'g_two')
plt.title("Histogram of g_one and g_two")
plt.legend()
```

```python
# can you make g_one red and g_two yellow?
```

```python
fig, ax = plt.subplots()
plt.hist(g_one, color ="red", label = 'g_one')
plt.hist(g_two, color = "yellow", label = 'g_two')
plt.title("Histogram of g_one and g_two")
plt.legend()
```

```python
# can you remove the colors and draw the average line for both
```

```python
fig, ax = plt.subplots()
plt.hist(g_one, label = 'g_one')
plt.hist(g_two, label = 'g_two')
plt.axvline(g_one.mean(), color='k', linestyle='dashed', linewidth=1, label = "g_one avg")
plt.axvline(g_two.mean(), color='g', linestyle='dashed', linewidth=1, label = "g_two avg")
plt.title("Histogram of g_one and g_two")
plt.legend()
```
