```python
%matplotlib inline
import pylab as plt
import numpy as np

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
```

```python
# plot y as a function of x in red color with title
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, "r")
plt.title("y as a function of x")
```

```python
# Now give me a histogram of the values of y as a function of x
```

```python
fig, ax = plt.subplots()
plt.bar(x, y)
plt.title("y as a function of x")
```

```python
# plot the line and histogram on one graph 
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, "r")
plt.bar(x, y)
plt.title("y as a function of x")
```
