```python
%matplotlib inline
import pylab as plt
import numpy as np

xs = [1,3,5,7,9]
ys = [x**2 for x in xs]
zs= [y**2 for y in ys]
```

```python
#plot xs function vs zs with title
```

```python
fig, ax = plt.subplots()
plt.plot(xs, zs)
plt.title("xs as a function of zs")
```

```python
#add ys as a function of zs with title and legend
```

```python
fig, ax = plt.subplots()
plt.plot(xs, zs, label = "zs")
plt.plot(xs, ys, label = "ys")
plt.legend()
plt.title("ys as a function of zs")
```

```python
#show the graphs along with the points
```

```python
fig, ax = plt.subplots()
plt.plot(xs, zs, "-o", label = "zs")
plt.plot(xs, ys, "-o", label = "ys")
plt.legend()
plt.title("ys as a function of zs")
```
